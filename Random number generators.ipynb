## Define a function to implement the above pseudo random number generator 

```python
def lcg(M, a, b, N0, k):
    """"
    Function that implements a linear congruential generator
    
    Input: 
          M     -> Given int
          a     -> Given int
          b     -> Given int
          N0    -> Seed
          k     -> Dimension
          
    Output:
         U      -> Pseudo random numbers
    """
    U = np.zeros(k)
    U[0] = N0/M
    for i in range(1, k):
        N0 = (a * N0 + b) % M
        U[i] = N0 / M
    return U
```

Visualise the pseudo random numbers $Z_i^j$ in $[−4, 4]$, the pairs $(Z_i^j, Z_{i+1}^j)$ in $[−4, 4]^2$ and the triples $(Z_i^j, Z_{i+1}^j, Z_{i+2}^j)$ in $[−4, 4]^3$ for $j = 1, 2$.


## Define a function to implement Marsaglia's polar algorithm

```python
def m_polar(U1, U2, k):
    """
    Function that implements Marsaglia's polar algorithm
    
    Input:
        U1 -> Uniform deviates 
        U2 -> Uniform deviates
        k  -> length of above RV
    
    Output:
        Z1 -> Normal deviates
        Z2 -> Normal deviates
        k_new -> length of above RV
    """
    
    V1 = 2 * U1 - 1
    V2 = 2 * U2 - 1
    Z1 = []
    Z2 = []

    for i in range(k):
        W = V1[i]*V1[i] + V2[i]*V2[i]

        if (W < 1 and W > 0):
            Q = sqrt(-2 * log(W) / W)
            Z1.append(V1[i] * Q)
            Z2.append(V2[i] * Q)
    k_new = len(Z1)
    Z1 = np.array(Z1)
    Z2 = np.array(Z2)
    return Z1, Z2, k_new
```
